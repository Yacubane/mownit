```julia
using Roots
using ForwardDiff
using Plots

f = Function[]
fstr = String[]
push!(fstr,"2 * x -exp(-x)")
push!(fstr,"x * exp(-x)")
push!(fstr,"( x + 3 ) * ( x - 1 )^2")
push!(fstr, "x^3")
push!(fstr,"cos(x) - x")
push!(fstr,"x^3 - 2*x - 5")

push!(f,x -> 2 * x -exp(-x))
push!(f,x -> x * exp(-x))
push!(f,x -> ( x + 3 ) * ( x - 1 )^2 )
push!(f,x -> x^3)
push!(f,x -> cos(x) - x)
push!(f,x -> x^3 - 2*x - 5)



D = Function[]
z_bisection = Float64[]
for fun in f
    push!(D,x -> ForwardDiff.derivative(fun, float(x)))
end

#fun(function, derrivative)
function evaluate_foreach(fun)
    for i in 1:length(f)
        println("Finding zero for function: ", fstr[i])
        zero = fun(i, f[i], D[i])
        println("f(" , zero, ") = ", f[i](zero))
        println("iszero = ", iszero(f[i](zero)))    
        println()
        println("=================================================================================")
        println()

    end
end
```

```julia
#Metoda bisekcji
fz_bisection = Float64[]
function zero(i, fun,der)
    zero_ = find_zero(fun,(-4,4), Bisection(),verbose=true)
    push!(fz_bisection, f[i](zero_))
    return zero_
end
evaluate_foreach(zero)
```

```julia
#Metoda Newtona
fz_newton= Float64[]
function zero(i, fun,der)
    zero_ = find_zero((fun, der),-1, Roots.Newton(),verbose=true)
    push!(fz_newton, f[i](zero_))
    return zero_
end
evaluate_foreach(zero)
```

```julia
#Metoda Steffensena 
fz_order2 = Float64[]
function zero(i, fun,der)
    zero_ = find_zero(fun,-1, Order2(),verbose=true)
    push!(fz_order2, f[i](zero_))
    return zero_
end
evaluate_foreach(zero)
```

```julia
# ZAD 3B
# https://en.wikipedia.org/wiki/Newton_fractal

using Images

fn(z) = z^3 - 1 
fn_der(z) = 3z^2
roots = [
    (1 + 0im),
    (-.5 + sqrt(3)/2im),
    (-.5 - sqrt(3)/2im)]

imagew = 512
imageh = 512
maxIteration = 200
tolerance = 0.000001;
znextfun = z -> z - (fn(z) / fn_der(z))
```

```julia
function color_fun(iteration, root_num)
    colors = [
    RGB(1, 0, 0),
    RGB(0, 1, 0),
    RGB(0, 0, 1),
    RGB(0, 0, 0)]
    if root_num == 0
        return RGB(0,0,0)
    end
    return colors[root_num]
end    
```

```julia
function color_fun(iteration, root_num)
    maxIteration_ = 20
    iteration = clamp(iteration, 1, maxIteration_)
    if(root_num == 1) return RGB(iteration/maxIteration_,0,0) end
    if(root_num == 2) return RGB(0,iteration/maxIteration_,0) end
    if(root_num == 3) return RGB(0,0,iteration/maxIteration_) end
    return RGB(1,0,0)
end    
```

```julia
function get_color(x, y, znextfun)
    zx = -2.5 + 3.5  * ((x-1) / (imagew-1))
    zy = -1 + 2 * ((y-1) / (imageh-1))
    z = zx + zy * im
    for iteration in 1:maxIteration
        z = znextfun(z)
        for j=1:length(roots)
            difference = z - roots[j]
            if abs(difference.re) < tolerance && abs(difference.im) < tolerance
                return color_fun(iteration,j)
            end
        end
    end        
    return color_fun(0,0)
end

image = zeros(RGB, 512 ,512)
for x=1:imagew
    for y=1:imageh
       image[y,x] = get_color(x,y, znextfun)
    end
end

image
```
